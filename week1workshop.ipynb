```python
#4
#For the following arrays:
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);
#Prove following with Numpy:
#Prove A*A^−1 = I.
invA = np.linalg.inv(A)
K = np.dot(A,invA)
print("A * A inverse =\n", K)
```

```python
#Prove AB ̸= BA.
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB =\n", AB)
print("BA =\n", BA)
```

```python
#Prove (AB)^T = B^T A^T
transposeAB = np.transpose(AB)
tranBtranA= np.dot(np.transpose(B), np.transpose(A))
print("Transpose of (AB)^T:\n", transposeAB)
print("Transpose of B^TA^T:\n", tranBtranA)
print("hence, (AB)^T = B^T A^T")
```

```python
#Solve the following system of Linear equation using Inverse Methods.{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
#2x − 3y + z = −1
#x − y + 2z = −3
#3x + y − z = 9
#Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}
A = np.array([[ 2, -3,  1],[ 1, -1,  2],[ 3,  1, -1]])
B = np.array([-1, -3, 9])
inverse= np.linalg.inv(A)
y = np.dot(inverse, B)
print("inverse:\n", y)
```
